# Visualización de datos integrada de Pandas

En esta practica, aprenderemos sobre las capacidades integradas de Pandas para la visualización de datos. Está construido a partir de matplotlib, ¡pero se convirtió a Pandas para un uso más fácil!

## Importaciones

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

## Los Datos

Hay algunos archivos csv de datos falsos que puede leer como dataframes:

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

## Hojas de estilo

Matplotlib tiene [style sheets](http://matplotlib.org/gallery.html#style_sheets) que puede usar para hacer que sus gráficos se vean un poco más bonitos. Estas hojas de estilo incluyen plot_bmh, plot_fivethirtyeight, plot_ggplot y más. Básicamente, crean un conjunto de reglas de estilo que siguen sus gráficos. Recomiendo usarlos, hacen que todas las graficas tengan el mismo aspecto y se sientan más profesionales. Incluso puedse crear el tuyo propio si deseas que los gráficos tengan el mismo aspecto.

A continuación se explica cómo utilizarlos.

**Antes de plt.style.use(), los gráficos se ven así:**

```python
df1['A'].hist()
```

Llamar al estilo:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Ahora los graficos se ven así:

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

¡Sigamos con el estilo ggplot y veamos cómo utilizar las capacidades de graficas integradas de pandas!

# Tipos de Graficas

Hay varios tipos de gráficos integrados en Pandas, la mayoría de ellos gráficos estadísticos por naturaleza:

* df.plot.area       
* df.plot.density  
* df.plot.hist          
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      


También puede simplemente llamar a df.plot (kind = 'hist') o reemplazar ese tipo de argumento con cualquiera de los términos clave que se muestran en la lista anterior (por ejemplo, 'box', 'barh', etc.)
___

Revisemos cada una de ellas.

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histogramas

```python
df1['A'].plot.hist(bins=50)
```

## Scatter Plots (Gráfico de dispersión)

```python
df1.plot.scatter(x='A',y='B')
```

Puede usar c para colorear en base a otro valor de columna. Usar cmap para indicar el mapa de colores a usar. Para todos los mapas de colores, consultar:: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

O use s para indicar el tamaño basado en otra columna. El parámetro s debe ser una matriz, no solo el nombre de una columna:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## BoxPlots (Diagrama de Cajas)

```python
df2.plot.box() #También puede pasar un argumento by = para groupby
```

## Hexagonal Bin Plot (Diagrama contenedor hexagonal)

Útil para datos bivariados, alternativa al diagrama de dispersión (scatter plot):

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```
