#### 1. ¿Existe diferencia entre la edad de los afiliados según el sexo? ¿Cuáles son las conclusiones? (2 puntos)
###### Sugerencia: Tabla de contingencia y gráfico de barras

```python
from cProfile import label
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('OPENDATA_AFILIADOS_DS01_202405_MUESTRA.csv')

fig, ax = plt.subplots(figsize=(24, 8))
pd.crosstab(df['EDAD'], df['SEXO']).plot(kind='bar', title='Distribucion de edad por sexo', ax=ax)

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

**Sí, la distribución de edad por sexo muestra que ambos sexos tienen patrones similares, pero hay diferencias en algunos rangos de edad: en la juventud adulta hay más afiliadas mujeres, mientras que en la adultez media y mayor la diferencia se reduce ligeramente. Sin embargo, la tendencia general es parecida para ambos sexos.**

#### 2. Describir la correlación entre la edad y el total de afiliados. ¿Cuáles son las conclusiones? (2 puntos)
###### Sugerencia: Calcular el valor de la correlación y su respectivo gráfico

```python
correlation = df['EDAD'].corr(df['TOTAL_AFILIADOS'])
print(f"El valor de correlacion entre la edad y el total de afiliados es: {correlation}")

# graficar
plt.scatter(df['EDAD'], df['TOTAL_AFILIADOS'])
plt.title('Relacion entre edad y total de afiliados')
plt.xlabel('Edad')
plt.ylabel('Total de afiliados')
plt.show()
```

**El diagrama de dispersión muestra que la mayoría de los afiliados se concentran en edades jóvenes y medias, y que a mayor edad, el total de afiliados tiende a disminuir. Esto se verifica el valor que toma la correlación negativa: a mayor edad, menor número de afiliados.**

#### 3. Realizar una análisis descriptivo de la edad de los afiliados y hacer uso de la visualización más adecuada. ¿Cuáles son las conclusiones? (2 puntos)

```python
print(df['EDAD'].describe())
plt.figure(figsize=(12,6))
sns.histplot(df['EDAD'], kde=True)
plt.title('Distribución de la Edad de los Afiliados')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```


**El histograma muestra que la mayor parte de los afiliados se encuentra entre los 0 y 40 años, con picos en la adultez joven. Hay una larga cola hacia edades mayores, pero la frecuencia disminuye considerablemente después de los 50 años..**

#### 4. ¿Existe diferencia entre la dispersión de la edad de los afiliados según su plan de seguro? ¿Cuáles son las conclusiones? (2 puntos)
###### Sugerencia: Realizar el gráfico de cajas, scatterplot

```python
import seaborn as sns


plt.figure(figsize=(12, 6))
sns.boxplot(x='PLAN_DE_SEGURO', y='EDAD', data=df)
plt.title('Distribucion de edad segun el plan de salud')
plt.xlabel('Plan de salud')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.show()
```

**El diagrama de cajas por plan de salud muestra diferencias claras: algunos planes como SIS GRATUITO concentran afiliados más jóvenes, mientras que otros como SIS INDEPENDIENTE tienen una mediana de edad mayor y mayor dispersión. Hay diferencias notables en la edad según el plan de seguro.**

#### 5. ¿Existe diferencia entre la dispersión de la edad de los afiliados según el régimen de financiamiento? ¿Cuáles son las conclusiones? (2 puntos)
###### Sugerencia: Realizar el gráfico de cajas

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='REGIMEN_FINANCIAMIENTO', y='EDAD', data=df)
plt.title('Distribucion de edad segun el Regimen de Financiamiento')
plt.xlabel('Regimen de Financiamiento')
plt.ylabel('Edad')
plt.xticks(rotation=45)
plt.show()
```

**El diagrama de cajas por regimen de financiamiento muestra que el régimen SUBSIDIADO tiene mayor dispersión y una mediana menor que el SEMICONTRIBUTIVO, que agrupa afiliados de mayor edad y menor dispersión, estas son las principales diferencias en la dispersión de la edad según el régimen de financiamiento**
